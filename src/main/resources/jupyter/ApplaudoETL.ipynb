### Fetch the data with a proper schema and filter files by name

* The `getFilesNames` function return the list of files inside a directory (in this case Azure Blob Storage container) that have a higher numeric value of the `fileNumber` parameter.
E.g. If `fileNumber` = 2 and there exists these files in the directory: 01.csv, 02.csv, 03.csv and 04.csv
The output will be an Array with 03.csv and 04.csv   
  * Notes: This function also could be re-written in a way to use the creation date of the files instead of the names.

* The `getDataFromBlobStorage` function fetches Products data from Azure Blob Storage and returns a DataFrame with a proper schema.  
A list of files are fetched according to the `fileNumber` parameter, by default fileNumber = -1, this means that all data in the directory will be returned.
By providing a value for `fileNumber` parameter, new files can be consumed from the Directory without repeat those already obtained in previous executions. This works by assuming that the new data will arrive in files with higher values names.

```scala
case class Product(order_id: Long, user_id: Long, order_number: Int, order_dow: Int, order_hour_of_day: Int,
                   days_since_prior_order: Float, order_detail: String)

val ProductSchema = Encoders.product[Product].schema

def getFileNames(path: String, fileNumber: Int): Array[String] = {
    val fs = FileSystem.get(new java.net.URI(path), new Configuration())
    val status = fs.listStatus(new Path(path))
    status.filter(x => x.getPath.toString.split("/").last.split("\\.").head.toInt > fileNumber)
      .map(x => x.getPath.toString)
}

def getDataFromBlobStorage(spark: SparkSession, fileNumber: Int = -1): DataFrame = {
    val path = s"wasbs://$ContainerName@$StorageAccountName.blob.core.windows.net"
    spark.read.csv(s"$path/00.csv")
    val listFiles = getFileNames(path, fileNumber)

    spark.read.schema(ProductSchema)
      .option("header", "false")
      .option("escape", "\"")
      .option("mode", "DROPMALFORMED").csv(listFiles: _*)
}

val dfProductsAzure = getDataFromBlobStorage(spark)
val rowsFullData = dfProductsAzure.count()
val rowsPartialData = getDataFromBlobStorage(spark,3).count()

dfProductsAzure.show(10)
println(s"Retrieve all the data. Number of rows: $rowsFullData")
println(s"Retrieve only data in files with a name higher than `03`. Number of rows: $rowsPartialData")
println(s"Proper schema of the data: ")
dfProductsAzure.schema.fields.foreach(f => println(s"\t\tName: ${f.name}, DataType: ${f.dataType}"))
```

### Fetch the data with a proper schema and filer rows by order_id

The `getDataFromSQLServer` function fetches Products data from a MSSQL Server instance and returns a DataFrame with a proper schema.
The data is fetched using a custom query that compares the field `order_id`, by default orderId = -1, this means that all data in the table will be returned. 
By providing a value for `order_id` field, new records can be consumed from the SQL Server table without repeat those already obtained in previous executions, since each `order_id` field is unique and it is assumed that the new data will have higher values.

```scala
def getDataFromSQLServer(spark: SparkSession, orderId: Int = -1): DataFrame = {

    val query = s"(select * from $dbTable where CAST(order_id AS int) > $orderId) as t"

    val df = spark.read
      .format("jdbc")
      .option("driver", driver)
      .option("url", mssqlUrl)
      .option("dbtable", query)
      .option("user", user)
      .option("password", password)
      .load()

    ProductSchema.fields.foldLeft(df) {
      (df, s) => df.withColumn(s.name, df(s.name).cast(s.dataType))
    }
}

val dfProductsMssql = getDataFromSQLServer(spark)
val rowsFullDataMssql = dfProductsMssql.count()
val rowsPartialDataMssql = getDataFromSQLServer(spark,146023).count()

dfProductsMssql.show(10)
println(s"Retrieve all rows from MSSQL. Number of rows: $rowsFullDataMssql")
println(s"Retrieve only rows in the table with a `order_id` higher than `146023`. Number of rows: $rowsPartialDataMssql")
println(s"Proper schema of the data: ")
dfProductsMssql.schema.fields.foreach(f => println(s"\t\tName: ${f.name}, DataType: ${f.dataType}"))
```

The `getDataFromAPI` function fetches Product Details from an API and returns a DataFrame with a proper schema. 
The function parse JSON data to a list of the class `ProductDetail`, and then convert the list into a DataFrame.   
This approach works because the data received from the API is small enough, but, in other circunstances this function could be re-written using libreries like [Rest Data Source for Apache Spark](https://github.com/sourav-mazumder/Data-Science-Extensions/tree/master/spark-datasource-rest)

```scala
org.apache.spark.sql.catalyst.encoders.OuterScopes.addOuterScope(this)

case class ProductDetail(product_name: String, aisle: String, department: String)

def getDataFromAPI(spark: SparkSession): DataFrame = {
    val url = "https://etlwebapp.azurewebsites.net/api/products"
    implicit val formats = DefaultFormats
    val data = List(scala.io.Source.fromURL(url).mkString)(0)
    val listMap = (parse(data) \\ "results" \\ "items").extract[List[Map[String,String]]]
    val listProductDetail = listMap.map(obj => ProductDetail(obj("product_name"),obj("aisle"),obj("department")))
    // For implicit conversions like converting RDDs to DataFrames
    import spark.implicits._
    listProductDetail.toDF()
}

val dfProductDetails = getDataFromAPI(spark)
val rowsProductDetails = dfProductDetails.count()
dfProductDetails.show(10)
println(s"Number of rows: $rowsProductDetails")
println(s"Schema of the data: ")
dfProductDetails.schema.fields.foreach(f => println(s"\t\tName: ${f.name}, DataType: ${f.dataType}"))
```

## Merge Product's data from Azure Blob Storage and MSSQL Server

The `mergeAndTransformProductData` function merge data from Azure Blob Storage and MSSQL sources. Clean the fields with bad data format. Also, split and explode the original `order_detail` column in multiples ones to accomplish the desired requirements.
* Notes:
 * Clean the "product" column that has non-ascii characters
 * Cast the "days_since_prior_order" column from Double to Int
 * Transformed unexpected data from the "order_hour_of_day" column

```scala
def mergeAndTransformProductData(dataFromBlob: DataFrame, dataFromSQLServer: DataFrame): DataFrame = {
    dataFromBlob.union(dataFromSQLServer)
      .withColumn("order_detail_array", split(col("order_detail"), "~"))
      .withColumn("order_detail_exploded", explode(col("order_detail_array")))
      .withColumn("product", split(col("order_detail_exploded"), "\\|").getItem(0))
      .withColumn("aisles", split(col("order_detail_exploded"), "\\|").getItem(1))
      .withColumn("number_of_products", split(col("order_detail_exploded"), "\\|").getItem(2).cast(IntegerType))
      .drop("order_detail", "order_detail_array", "order_detail_exploded")
      .withColumn("product", regexp_replace(col("product"), "[^\\x00-\\x7F]", ""))
      .withColumn("days_since_prior_order", col("days_since_prior_order").cast(IntegerType))
      .withColumn("order_hour_of_day", when(col("order_hour_of_day") === 24, 0).
        otherwise(col("order_hour_of_day")))
}

val dfProducts = mergeAndTransformProductData(dfProductsAzure,dfProductsMssql)
val rowsProducts = dfProducts.count()

dfProducts.show(10)
println(s"Number of rows: $rowsProducts")
println(s"Schema of the data: ")
dfProducts.schema.fields.foreach(f => println(s"\t\tName: ${f.name}, DataType: ${f.dataType}"))
```

## Clean and validate data values

Some of the fields in the DataFrame have values that do not correctly represent their meaning. E.g. the `order_hour_of_day`, in principle this field should represent values in the range 0-23.

```scala
dfProductJoined.select("order_hour_of_day").filter(column("order_hour_of_day") < 0).show(10)
```

To solve this problem in a generic way, the `validateDataValues` function takes care of transform all columns in the DataFrame.
* Notes:
Notes:
 * Numeric columns cannot be < 0
 * Delete white-spaces from string columns

```scala
def validateDataValues(dfProduct: DataFrame): DataFrame = {
    dfProduct.schema.fields.foldLeft(dfProduct) {
      (df, column) => {
        column.dataType match {
          case _: StringType => df.withColumn(column.name, trim(df(column.name)))
          case _: IntegerType => df.withColumn(column.name, abs(df(column.name)))
          case _: LongType => df.withColumn(column.name, abs(df(column.name)))
          case _: FloatType => df.withColumn(column.name, abs(df(column.name)))
          case _: DoubleType => df.withColumn(column.name, abs(df(column.name)))
          case _ => df.withColumn(column.name, df(column.name))
        }
      }
    }
}

val dfValidated = validateDataValues(dfProductJoined)

dfValidated.select("order_hour_of_day").filter(column("order_hour_of_day") < 0).show(10)

println("After validate, there's no data with an invalid range of values")
```

* The `writoIntoBigQuery` function store de DataFrame into a BigQuery dataset and the `tableName` parameter allows storing new data from different executions in different tables.

```scala
def writoIntoBigQuery(df: DataFrame, datasetName: String, tableName:String) {
    df.write.mode("overwrite").format("bigquery")
      .option("table", s"$datasetName.$tableName")
      .save()
}


def storeData(df: DataFrame, resultPath: String): Unit = {
    df.write.mode("overwrite").parquet(s"$resultPath")
}

val rowsValidatedProducs = dfValidated.count()


println(s"Number of rows: $rowsValidatedProducs")
println(s"Schema of the final table of Products: ")
dfValidated.schema.fields.foreach(f => println(s"\t\tName: ${f.name}, DataType: ${f.dataType}"))
```

## Clients category and segmentation for Marketing and Customer Loyalty Departments 

### Clients category

To solve this problem the `getClientsCategory` function uses a Window to add the next:
* `total_products_bought` column: All the products that a user bought
* `mom_products` column: All products in the Mom's category bought by a user
* `single_products` column: All products in the Single category bought by a user
* `pet_friendly_products` column: All products in the Pet Frienldy category bought by a user   

After adding all those columns, they're used like parameters for a UDF that classify every client. 

```scala
  def getClientsCategory(dfProducts: DataFrame): DataFrame = {
    val momItems = List("dairy eggs", "bakery", "household", "babies")
    val singleItems = List("canned goods", "meat seafood", "alcohol", "snacks", "beverages")
    val petFriendlyItems = List("canned goods", "pets", "frozen")

    // UDF to Classify clients according to the established rules
    val clientsCategoryUdf = udf((total: Int, mom: Int, single: Int, petFriendly: Int) => {
      var category: String = "A complete mystery"
      if (mom / total > 0.5) {
        category = "Mom"
      } else if (single / total > 0.6) {
        category = "Single"
      } else if (petFriendly / total > 0.3) {
        category = "Pet Friendly"
      }
      category
    })

    val window = Window.partitionBy("user_id")

    dfProducts.withColumn("total_products_bought", sum("number_of_products").over(window))
      .withColumn("mom_products", sum(when(col("department").isin(momItems: _*),
        col("number_of_products")).otherwise(0)).over(window))
      .withColumn("single_products", sum(when(col("department").isin(singleItems: _*),
        col("number_of_products")).otherwise(0)).over(window))
      .withColumn("pet_friendly_products", sum(when(col("department").isin(petFriendlyItems: _*),
        col("number_of_products")).otherwise(0)).over(window))
      .withColumn("category", clientsCategoryUdf(col("total_products_bought"), col("mom_products"),
        col("single_products"), col("pet_friendly_products")))
      .select("user_id", "category").dropDuplicates("user_id")
}



val dfCategory = getClientsCategory(dfValidated)

dfCategory.show(10)
dfCategory.groupBy("category").count().show()
```

### Clients segmentation

To solve this problem the `getClientsSegmentation` function uses a Window to add the next:

* `total_products_bought` column: All the products that a user bought

Also, it's necessary to calculate the Quartiles of every day of the week. To do that, it's used the `approxQuantile` stat function. The results of those calculations are stored in a Map.   
Finally, to get the Clients Segmentation it's applied the `clientSegmentUdf` UDF. 

```scala
import scala.collection.mutable

def getClientsSegmentation(dfProducts: DataFrame): DataFrame = {

    // UDF to Segment clients according to the established rules.
    def clientsSegmentUdf(m: mutable.Map[(String, Int), Double]) = udf((orderDow: Int, dspo: Int,
                                                                        totalProductsBought: Int) => {
      var segment: String = "Undefined"
      if (dspo <= 7 && totalProductsBought > m("third", orderDow)) {
        segment = "You've Got a Friend in Me"
      } else if ((dspo >= 10 && dspo <= 19) && totalProductsBought > m("second", orderDow)) {
        segment = "Baby come Back"
      } else if (dspo > 20 && totalProductsBought > m("first", orderDow)) {
        segment = "Special Offers"
      }
      segment
    })

    // The Quartiles of every day are calculated and passed like param to the Client Segment UDF
    // The function approxQuantile is used with 0.0 in the third argument, this way the results are exact and not
    // approximations
    val quartileMap = mutable.Map.empty[(String, Int), Double]
    for (day <- (0 to 6).toList) {
      val medianAndQuantiles = dfProducts.filter(col("order_dow") === day).stat.approxQuantile("number_of_products",
        Array(0.25,0.5, 0.75), 0.0)
      quartileMap(("first", day)) = medianAndQuantiles(0)
      quartileMap(("second", day)) = medianAndQuantiles(1)
      quartileMap(("third", day)) = medianAndQuantiles(2)
    }

    val window = Window.partitionBy("user_id")
    dfProducts.withColumn("total_products_bought", sum("number_of_products").over(window))
      .withColumn("client_segment", clientsSegmentUdf(quartileMap)(col("order_dow"),
        col("days_since_prior_order"), col("total_products_bought")))
      .select("user_id", "client_segment").dropDuplicates("user_id")
}


val dfSegmentation = getClientsSegmentation(dfValidated)

dfSegmentation.show(10)
dfSegmentation.groupBy("client_segment").count().show()
```
